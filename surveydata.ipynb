```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url_2021 = 'https://github.com/UVADS/orientation-technical/raw/main/dat/2021-06-15-survey.csv'
df_2021 = pd.read_csv(url_2021)
print(df_2021.head(5))
```

```python
plt.hist(df_2021['CPU Cycle Rate (in GHz)'], bins = 15)
plt.xlabel('CPU Cycle Rate (in GHz)')
plt.ylabel('Number of Students')
plt.title('2021 Cohort Data')
plt.show()
```

```python
plt.hist(df_2021['CPU Number of Cores'])
plt.xlabel('CPU Number of Cores')
plt.ylabel('Number of Students')
plt.title('2021 Cohort Data')
plt.show()
```

```python
df_2021 = df_2021[df_2021['RAM (in GB)'] < 100]
plt.hist(df_2021['RAM (in GB)'])
plt.xlabel('RAM (in GB)')
plt.ylabel('Number of Students')
plt.title('2021 Cohort Data')
plt.show()
```

```python
plt.hist(df_2021['Hard Drive Size (in GB)'], bins = 6)
plt.xlabel('Hard Drive Size (in GB)')
plt.ylabel('Number of Students')
plt.title('2021 Cohort Data')
plt.show()
```
