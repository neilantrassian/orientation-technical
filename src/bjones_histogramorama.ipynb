```python
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv("https://raw.githubusercontent.com/UVADS/orientation-technical/main/dat/2022-06-14-survey.csv")
data_filtered=data[data["CPU Number of Cores (int)"]>0]


ram_data=data_filtered["RAM (in GB)"]
plt.hist(ram_data, bins=15)
plt.title("RAM (in GB)")
plt.show()
```

```python
vars=data_filtered.columns
for var in vars:
    plt.hist(data_filtered[var])
    plt.title(var)
    plt.show()
```

Is there a relationship between hard drive size and operating system?

```python
os=data_filtered["Operating System"]
hd_size=data_filtered["Hard Drive Size (in GB)"]
plt.scatter(os, hd_size, color="r", alpha=.4)
plt.show()
```

```python
import seaborn as sns
sns.stripplot('Operating System', 'Hard Drive Size (in GB)', data=data_filtered, jitter=0.2, alpha=.5)
sns.despine()
```
