```python
import pandas as pd
survey2 = pd.read_csv("2022-06-14-survey.csv")

# removes outliers with less than 1 cores
for i in range(0, len(survey2.index)):
    if (survey2["CPU Number of Cores (int)"][i] < 1):
        survey2 = survey2.drop(i)
        
hist = survey2.hist(bins=10, figsize=(20, 10), color = "gold", ec="green", grid = False)
```
