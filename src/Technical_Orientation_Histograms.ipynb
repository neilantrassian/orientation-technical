### Technical Orientation: Histograms

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df_21 = pd.read_csv("https://raw.githubusercontent.com/UVADS/orientation-technical/main/dat/2021-06-15-survey.csv")
df_21;
```

```python
df_22 = pd.read_csv("https://raw.githubusercontent.com/UVADS/orientation-technical/main/dat/2022-06-14-survey.csv")
df_22;
```

```python
os_21 = df_21['Operating System'].to_numpy()
cpu_21 = df_21['CPU Cycle Rate (in GHz)'].to_numpy()
ram_21 = df_21['RAM (in GB)'].to_numpy()
hd_21 = df_21['Hard Drive Size (in GB)'].to_numpy()
```

```python
os_21 = [x.replace(' 10', '') for x in os_21]
ram_21 = [x for x in ram_21 if x < 100]
```

```python
os_22 = df_22['Operating System'].to_numpy()
cpu_22 = df_22['CPU Cycle Rate (in GHz)'].to_numpy()
ram_22 = df_22['RAM (in GB)'].to_numpy()
hd_22 = df_22['Hard Drive Size (in GB)'].to_numpy()
```

```python
ram_22 = [x for x in ram_22 if x < 100]
cpu_22 = [x for x in cpu_22 if x < 400]
```

```python
fig, ax = plt.subplots(2,2, figsize = (16,14))
fig.subplots_adjust(wspace=0.2, hspace=0.2, left = 0.2, right = None)

ax[0,0].hist([os_21, os_22], color = ['lightblue', 'orange'], alpha = 0.5, stacked = False, label = ['2021', '2022'])
ax[0,0].set_title('Operating System')
ax[0,0].legend()

ax[0,1].hist([cpu_21, cpu_22], color = ['lightblue', 'orange'], alpha = 0.5, stacked = False, label = ['2021', '2022'])
ax[0,1].set_title('CPU Cycle Rate (in GHz)')
ax[0,1].legend()

ax[1,0].hist([ram_21, ram_22], color = ['lightblue', 'orange'], alpha = 0.5, stacked = False, label = ['2021', '2022'])
ax[1,0].set_title('RAM (in GB)')
ax[1,0].legend()

ax[1,1].hist([hd_21, hd_22], color = ['lightblue', 'orange'], alpha = 0.5, stacked = False, label = ['2021', '2022'])
ax[1,1].set_title('Hard Drive Size (in GB)')
ax[1,1].legend();
```
