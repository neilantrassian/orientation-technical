```python
# Import datasets 
import pandas as pd
lastyear = pd.read_csv('https://raw.githubusercontent.com/UVADS/orientation-technical/main/dat/2021-06-15-survey.csv')
thisyear = pd.read_csv('https://raw.githubusercontent.com/UVADS/orientation-technical/main/dat/2022-06-14-survey.csv')
```

```python
# Select column with information about the RAM (exclude any outliers > 32 GB)

# Dataset from 2021
x1 = lastyear['RAM (in GB)']
x1 = x1[x1<33]

# Dataset from 2022
x2 = thisyear['RAM (in GB)']
x2 = x2[x2<33]
```

```python
# Plot histogram
import matplotlib.pyplot as plt
import numpy as np

plt.figure(dpi=800)
plt.hist([x1,x2],bins=12,label=['2021','2022'],align='mid',color=['darkorange','midnightblue'])
plt.title('RAM of computers used by MSDS students')
plt.xlabel('RAM (GB)')
plt.xticks(ticks=[4,8,16,32])
plt.ylabel('Counts')
plt.legend()
plt.show()
```
